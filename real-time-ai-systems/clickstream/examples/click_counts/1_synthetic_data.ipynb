```python
import pandas as pd
import numpy as np
import hopsworks

n_rows = 100_000
n_users = 100
time_window_hours = 1

user_ids = np.arange(1, n_users + 1)

np.random.seed(42)

user_id_col = np.random.choice(user_ids, size=n_rows, replace=True)


start_time = pd.Timestamp.now().floor("H")
end_time = start_time + pd.Timedelta(hours=time_window_hours)

event_time_col = np.random.randint(
    int(start_time.timestamp()),   # start as seconds
    int(end_time.timestamp()),     # end as seconds
    size=n_rows
)

click_id_col = np.arange(1, n_rows + 1)

df = pd.DataFrame({
    "user_id": user_id_col,
    "event_time": event_time_col.astype("int64"),
    "click_id": click_id_col
})
df
```

```python
project = hopsworks.login()
fs = project.get_feature_store()

clicks_fg = fs.get_or_create_feature_group(
    name="user_clicks",
    version=1,
    # topic_name=f"{project.name}_user_clicks",
    description="Synthetic user clickstream data (100k rows, 100 users, 24h window)",
    primary_key=["user_id", "event_time"],
    event_time="event_time",
    online_enabled=True,
)


print("✅ Created Feature Group: user_clicks v1")
clicks_fg.insert(df)
print("✅ Data written to Hopsworks Feature Group")
```
