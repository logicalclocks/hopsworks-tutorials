# Read Features

```python
import hopsworks

project = hopsworks.login()
fs = project.get_feature_store()
```

```python
ctr_fg = fs.get_feature_group("ctr_5min_flink", version=1)
df = ctr_fg.read()

print(f"Total records: {len(df)}")
df.sort_values("window_end", ascending=False).head(10)
```

```python
print(f"Average CTR: {df['ctr'].mean():.2%}")
print(f"Median CTR: {df['ctr'].median():.2%}")
```

```python
fv = fs.get_or_create_feature_view(
    name="ctr_flink_fv",
    version=1,
    query=ctr_fg.select_all()
)

fv.init_serving()

# Lookup specific user
features = fv.get_feature_vector({"user_id": "user_001"})
if features:
    print(f"CTR: {features[3]:.2%}")
```
